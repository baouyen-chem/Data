**Example 2** Develop a class Calculator with methods to add and subtract two numbers.

```python
class Calculator_1:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def add(self):
        return self.a + self.b
    def subtract(self):
        return self.a - self.b
p1 = Calculator_1()
print("sum: ", p1.add(5,3))
print("subtract: ", p1.subtract(6, 3))
```
