# While Loops


Print i as long as i is less than 6:

```python
i = 1 
if i < 6: 
    print(i)
    i += 1
```

Exit the loop when i is 3:

```python
i = 1
while i < 6: 
    print(i)
    if i == 3:
        break 
    i += 1
#i += 1 must be at the end 
```

Continue to the next iteration if i is 3 (pass the current loop)

```python
i = 0 
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

Print a message once the condition is false:

```python
i = 1
while i < 6:
    print (i)
    i += 1
else:
    print("i is no longer less than 6")
```
