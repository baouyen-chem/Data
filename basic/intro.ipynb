# While Loops


Print i as long as i is less than 6:

```python
i = 1 
if i < 6: 
    print(i)
    i += 1
```

Exit the loop when i is 3:

```python
i = 1
while i < 6: 
    print(i)
    if i == 3:
        break 
    i += 1
#i += 1 must be at the end 
```

Continue to the next iteration if i is 3 (pass the current loop)

```python
i = 0 
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

Print a message once the condition is false:

```python
i = 1
while i < 6:
    print (i)
    i += 1
else:
    print("i is no longer less than 6")
```

# Built-in Types

**List** 

```python
# Nhập list bằng vòng lặp và append()
n = int(input("Nhap so luong phan tu: "))
my_list = []
for i in range (n):
    item = input(f"Nhập phần tử thứ {i+1}: ")
    my_list.append(item)
print(my_list)

# Nhập list bằng map() và split()
a = list(map(int(input("Nhập từng số vào: ")))).split() 
# Ghi như này sai vì chuỗi->split->int(map)->list (ở đây map chỉ trả giá trị là iterator, cần iterable)
b = list(map(int, input("Nhap so lieu vao: ").split()))
print(b)
```
