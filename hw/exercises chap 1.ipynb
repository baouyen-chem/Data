**R-1.1** Write a short Python function, is multiple(n, m), that takes two integer
values and returns True if n is a multiple of m, that is,n = mi for some
integer i,and False otherwise.

```python
def multiple ():
    try:
        m = int(input("Enter m: "))
        n = int(input("Enter n: "))
        if n % m == 0:
            return True
        else:
            return False
    except ValueError:
        return "error"
    except ZeroDivisionError:
        return "ko chia 0"
print(multiple())
# Input ở dạng gì
# try...except giúp bắt lỗi mà không làm chương trình dừng đột ngột
def multiple_shorter(n: int, m: int) -> bool:
    return True if n % m == 0 else False
print(multiple_shorter(6,2))
print(multiple_shorter(2,6))
```

**R-1.3** Write a short Python function, minmax(data), that takes a sequence of
one or more numbers, and returns the smallest and largest numbers, in the
form of a tuple of length two. Do not use the built-in functions min or
max in implementing your solution.

```python
def minmax():
    a = list(map(int, input("Nhap so lieu vao va cach boi khoang trong: ").split()))
    imin = imax = a[0]
    for x in a: 
        if x < imin:
            imin = x
        if x > imax:
            imax = x
    print(f"Min: {imin}, Max: {imax}")
minmax()
```
